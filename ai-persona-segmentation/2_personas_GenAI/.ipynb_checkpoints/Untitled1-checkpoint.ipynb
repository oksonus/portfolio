```python
import pandas as pd
import glob
import os

# -----------------------------
# 1. PATH + FILE DISCOVERY
# -----------------------------
path = "C:/Users/okson/Downloads/Data Engineer Assessment Files/"
files = glob.glob(os.path.join(path, "Telephony*.xlsx"))

print(f"Files found: {len(files)}")

dfs = []

# -----------------------------
# 2. LOAD + TRANSFORM EACH FILE
# -----------------------------
for f in files:
    df = pd.read_excel(f)

    filename = os.path.basename(f)

    # -------------------------
    # Extract business date from filename
    # Example: Telephony 20200109.xlsx
    # -------------------------
    date_str = filename.replace("Telephony", "").replace(".xlsx", "").strip()
    file_dt = pd.to_datetime(date_str, format="%Y%m%d")

    # -----------------------------
    # Add metadata columns
    # -----------------------------
    df["date"] = file_dt
    df["file_date"] = file_dt
    df["source_file"] = filename

    dfs.append(df)

# -----------------------------
# 3. COMBINE ALL FILES
# -----------------------------
df = pd.concat(dfs, ignore_index=True)

# -----------------------------
# 4. CLEANING / STANDARDISATION
# -----------------------------

# Standardise agent names
df["agent"] = df["agent"].astype(str).str.strip().str.upper()

# Ensure numeric columns
numeric_cols = [
    "answered",
    "available_time",
    "talk_time",
    "hold_time",
    "wrap_time",
    "total_handle_time"
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0)

# -----------------------------
# 5. BUSINESS RULE CHECK (optional but good)
# -----------------------------
df["calculated_handle_time"] = df["talk_time"] + df["hold_time"]

df["handle_time_check"] = (
    df["calculated_handle_time"] == df["total_handle_time"]
)

# Keep only valid rows (optional)
df = df[df["handle_time_check"] == True].copy()

# -----------------------------
# 6. DEDUPLICATION (IDEMPOTENCY)
# -----------------------------
df = df.sort_values("date")

df = df.drop_duplicates(
    subset=["date", "agent"],
    keep="last"
)

# -----------------------------
# 7. DERIVED METRICS (ANALYTICS LAYER)
# -----------------------------

df["occupancy_rate"] = df["talk_time"] / df["available_time"]
df["aht"] = df["total_handle_time"] / df["answered"]

df["occupancy_rate"] = df["occupancy_rate"].fillna(0)
df["aht"] = df["aht"].replace([float("inf")], 0)

# -----------------------------
# 8. FINAL OUTPUT
# -----------------------------
print("Final dataset shape:", df.shape)
print(df.head())

# Optional: save output
df.to_csv("telephony_transformed.csv", index=False)
```

```python
import pandas as pd
import glob
import os
import re

# =========================
# 1. Folder path
# =========================
path = "your_folder_path/*"
files = glob.glob(path)

df_list = []

# =========================
# 2. Loop through files
# =========================
for file in files:

    filename = os.path.basename(file)

    # =========================
    # 3. Keep only Telephony files (case-insensitive)
    # =========================
    if "telephony" not in filename.lower():
        continue

    # =========================
    # 4. Read file safely (CSV or Excel)
    # =========================
    try:
        if file.lower().endswith(".csv"):
            df = pd.read_csv(file)
        elif file.lower().endswith((".xlsx", ".xls")):
            df = pd.read_excel(file)
        else:
            print(f"Skipping unsupported file type: {filename}")
            continue
    except Exception as e:
        print(f"Error reading {filename}: {e}")
        continue

    # =========================
    # 5. Extract 8-digit date from filename
    # =========================
    match = re.search(r"(\d{8})", filename)
    file_date = match.group(1) if match else None

    df["file_date"] = file_date

    # optional: convert to real date
    df["file_date"] = pd.to_datetime(df["file_date"], format="%Y%m%d", errors="coerce")

    df_list.append(df)

# =========================
# 6. Final safety check
# =========================
if not df_list:
    raise ValueError("No Telephony files found. Check folder path or filenames.")

# =========================
# 7. Append all
# =========================
final_df = pd.concat(df_list, ignore_index=True)

# =========================
# 8. Output
# =========================
print(final_df.head())
print("\nShape:", final_df.shape)
```

```python
import pandas as pd
import glob
import os

path = r"C:\Users\okson\Downloads\Data Engineer Assessment Files\*"
files = glob.glob(path)

df_list = []

telephony_files = []

print("FILES FOUND:")

# =========================
# 1. Collect Telephony files + dates
# =========================
for f in files:
    filename = os.path.basename(f)
    print(filename)

    if "Telephony" not in filename:
        continue

    digits = "".join([c for c in filename if c.isdigit()])
    file_date = digits[:8] if len(digits) >= 8 else None

    if file_date:
        telephony_files.append((f, file_date))

if not telephony_files:
    raise ValueError("No Telephony files found")

# =========================
# 2. Find newest date
# =========================
max_date = max([d for _, d in telephony_files])

print("\nNewest file date:", max_date)

# =========================
# 3. Load ONLY files < max_date (exclude newest)
# =========================
for file, file_date in telephony_files:

    if file_date >= max_date:   # <-- KEY CHANGE (exclude newest)
        continue

    df = pd.read_csv(file)
    df["file_date"] = file_date

    df_list.append(df)

# =========================
# 4. Combine safely
# =========================
if not df_list:
    raise ValueError("No files loaded (check date filtering)")

final_df = pd.concat(df_list, ignore_index=True)

print(final_df.head())
print("Shape:", final_df.shape)
```

```python
import pandas as pd
import glob
import os

path = r"C:\Users\okson\Downloads\Data Engineer Assessment Files\*"
files = glob.glob(path)

dim_df = None

# =========================
# 1. Load ONLY Agents file
# =========================
for file in files:
    filename = os.path.basename(file)

    if "Agents Team List" in filename:
        dim_df = pd.read_csv(file)
        break

if dim_df is None:
    raise ValueError("Agents file not found")

# =========================
# 2. Clean dates in dimension
# =========================
dim_df["date_from"] = pd.to_datetime(dim_df["date_from"], dayfirst=True, errors="coerce")
dim_df["date_to"] = pd.to_datetime(dim_df["date_to"], dayfirst=True, errors="coerce")

# open-ended end dates
dim_df["date_to"] = dim_df["date_to"].fillna(pd.Timestamp("2099-12-31"))

# =========================
# 3. Normalize keys
# =========================
final_df["agent_key"] = final_df["agent"].str.strip().str.upper()
dim_df["name_key"] = dim_df["employee_name"].str.strip().str.upper()

# =========================
# 4. Merge
# =========================
merged = final_df.merge(
    dim_df,
    left_on="agent_key",
    right_on="name_key",
    how="left"
)

# =========================
# 5. Apply SCD2 logic
# =========================
merged = merged[
    (merged["file_date"] >= merged["date_from"]) &
    (merged["file_date"] <= merged["date_to"])
]

# =========================
# 6. Cleanup
# =========================
merged = merged.drop(columns=["agent_key", "name_key"])

print(merged.head())
print("Shape:", merged.shape)
```

```python
import pandas as pd
import glob
import os

# =========================
# 1. PATH
# =========================
path = r"C:\Users\okson\Downloads\Data Engineer Assessment Files\*"
files = glob.glob(path)

print("FILES FOUND:")
for f in files:
    print(os.path.basename(f))

# =========================
# 2. TELEPHONY FILE PROCESSING
# =========================
telephony_files = []
df_list = []

for f in files:
    filename = os.path.basename(f)

    if "Telephony" not in filename:
        continue

    digits = "".join([c for c in filename if c.isdigit()])
    file_date = digits[:8] if len(digits) >= 8 else None

    if file_date:
        telephony_files.append((f, file_date))

if not telephony_files:
    raise ValueError("No Telephony files found")

# find newest file date
max_date = max([d for _, d in telephony_files])
print("\nNewest file date:", max_date)

# load all EXCEPT newest
for file, file_date in telephony_files:

    if file_date >= max_date:
        continue

    df = pd.read_csv(file)

    df["file_date"] = pd.to_datetime(file_date, format="%Y%m%d")

    df_list.append(df)

if not df_list:
    raise ValueError("No Telephony data loaded")

final_df = pd.concat(df_list, ignore_index=True)

# =========================
# 3. LOAD AGENTS FILE (FIXED)
# =========================
dim_file = None

for f in files:
    filename = os.path.basename(f)

    if "Agents Team List" in filename:
        dim_file = f
        break

if dim_file is None:
    raise ValueError("Agents file not found")

# FIXED READ (handles encoding + broken rows)
dim_df = pd.read_csv(
    dim_file,
    encoding="latin1",
    engine="python",
    on_bad_lines="skip"
)

# =========================
# 4. CLEAN DIMENSION
# =========================
dim_df["date_from"] = pd.to_datetime(dim_df["date_from"], dayfirst=True, errors="coerce")
dim_df["date_to"] = pd.to_datetime(dim_df["date_to"], dayfirst=True, errors="coerce")

# fill open-ended dates
dim_df["date_to"] = dim_df["date_to"].fillna(pd.Timestamp("2099-12-31"))

# =========================
# 5. NORMALISE KEYS
# =========================
final_df["agent_key"] = final_df["agent"].str.strip().str.upper()
dim_df["name_key"] = dim_df["employee_name"].str.strip().str.upper()

# =========================
# 6. MERGE
# =========================
merged = final_df.merge(
    dim_df,
    left_on="agent_key",
    right_on="name_key",
    how="left"
)

# =========================
# 7. SCD TYPE 2 FILTER
# =========================
merged = merged[
    (merged["file_date"] >= merged["date_from"]) &
    (merged["file_date"] <= merged["date_to"])
]

# =========================
# 8. CLEAN OUTPUT
# =========================
merged = merged.drop(columns=["agent_key", "name_key"])

print("\nFINAL OUTPUT:")
print(merged.head())
print("\nShape:", merged.shape)
```
